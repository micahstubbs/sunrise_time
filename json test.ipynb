```python
#!/usr/bin/env python
import datetime
import ephem # to install, type$ pip install pyephem

#define the function
def calculate_time(date, lat, long, is_rise, utc_time):
    o = ephem.Observer()
    o.lat, o.long, o.date = lat, long, date
    sun = ephem.Sun(o)
    next_event = o.next_rising if is_rise else o.next_setting
    return str(ephem.Date(next_event(sun, start=o.date) + utc_time*ephem.hour
                      ).datetime().strftime('%H:%M')) + " " + str(o.date.datetime().strftime('%D'))  
```
