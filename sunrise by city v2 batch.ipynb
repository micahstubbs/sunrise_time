```python
#!/usr/bin/env python
import datetime
import ephem # to install, type$ pip install pyephem

#define the function that returns the sunrise time
def calculate_time(date, lat, long, is_rise, utc_time):
    o = ephem.Observer()
    o.lat, o.long, o.date = lat, long, date
    sun = ephem.Sun(o)
    next_event = o.next_rising if is_rise else o.next_setting
    return str(ephem.Date(next_event(sun, start=o.date) + utc_time*ephem.hour
                      ).datetime().strftime('%H:%M')) + " " + str(o.date.datetime().strftime('%D'))  

#define the function that loops through a date range and calculates the sunrise time for each day
def sunrises(begin_date, end_date):
    for dt in rrule(DAILY, dtstart=a, until=b):
        n = calculate_time(dt, lat, long, is_rise, utc_offset) #calculate the sunrise time for each day in the range
        m.append(n) #add the sunrise time to a list
        
#define the function that loops through a date range and calculates the sunrise time for each day
def sunrises_dst(begin_date, end_date):
    for dt in rrule(DAILY, dtstart=a, until=b):
        n = calculate_time(dt, lat, long, is_rise, utc_dst_offset)
        m.append(n)
    
```
